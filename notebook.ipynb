### Cleaning Data

```python
# Drop the 'instant' and 'windspeed' columns from the day_df dataset.
del hour_df
```

```python
# Drop the 'instant' and 'windspeed' columns from the day_df dataset.
drop_column = ['instant','windspeed']

for i in day_df.columns:
    if i in drop_column:
        day_df.drop(i, axis=1, inplace=True)
day_df.head()
```

```python
# Rename the columns in the day_df dataset.
day_df.rename(columns={'dteday':'date', 'yr':'year', 'mnth':'month', 'weathersit':'weather_cond', 'hum':'humidity', 'cnt':'count'}, inplace=True)
day_df.head()
```

```python
# Convert the month, season, holiday, weekday, workingday, and weather_cond columns to categorical data type.
day_df['month'] = day_df['month'].map({
    1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun',
    7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'
})
day_df['season'] = day_df['season'].map({
    1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'
})
day_df['weekday'] = day_df['weekday'].map({
    0: 'Sun', 1: 'Mon', 2: 'Tue', 3: 'Wed', 4: 'Thu', 5: 'Fri', 6: 'Sat'
})
day_df['weather_cond'] = day_df['weather_cond'].map({
    1: 'Clear/Partly Cloudy',
    2: 'Misty/Cloudy',
    3: 'Light Snow/Rain',
    4: 'Severe Weather'
})
```

### Exploratory Data Analysis (EDA)

```python
day_df.info()
```

```python
# Convert the date column to datetime data type.
day_df['date'] = pd.to_datetime(day_df['date'])

day_df['season'] = day_df['season'].astype('category')
day_df['year'] = day_df['year'].astype('category')
day_df['month'] = day_df['month'].astype('category')
day_df['holiday'] = day_df['holiday'].astype('category')
day_df['weekday'] = day_df['weekday'].astype('category')
day_df['workingday'] = day_df['workingday'].astype('category')
day_df['weather_cond'] = day_df['weather_cond'].astype('category')

day_df.head()
```

```python
# Group bike rentals (both casual and registered) by month.
day_df.groupby(by='month', observed=True).agg({
    'count': ['max', 'min', 'mean', 'sum']
})
```

```python
# Group bike rentals (both casual and registered) by weather condition.
day_df.groupby(by='weather_cond', observed=True).agg({
    'count': ['max', 'min', 'mean', 'sum']
})
```

```python
# Group bike rentals (both casual and registered) by weekday.
day_df.groupby(by='weekday', observed=True).agg({
    'count': ['max', 'min', 'mean',]
})
```

```python
# Group bike rentals (both casual and registered) by holiday.
day_df.groupby(by='holiday', observed=True).agg({
    'count': ['max', 'min', 'mean']
})
```

```python
# Group bike rentals (both casual and registered) by workingday.
day_df.groupby(by='workingday', observed=True).agg({
    'count': ['max', 'min', 'mean']
})
```

### Visualization & Explanatory Analysis

#### Pertanyaan Bisnis 1: Apakah kondisi cuaca mempengaruhi jumlah pengguna sepeda?

```python
plt.figure(figsize=(10,6))
sns.barplot(
    x='weather_cond',
    y='count',
    data=day_df)

plt.title('Jumlah Pengguna Sepeda berdasarkan Kondisi Cuaca')
plt.xlabel('Kondisi Cuaca')
plt.ylabel('Jumlah Pengguna Sepeda')
plt.show()
```

###### Based on the bar plot above, it can be concluded that the number of bike users is higher when the weather is clear or partly cloudy. Meanwhile, the number of bike users is lower when the weather is misty or cloudy.

#### Pertanyaan Bisnis 2: Bagaimana perbedaan penggunaan sepeda pada hari kerja, hari libur, dan hari biasa?

```python
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(15,10))

# based on working days
sns.barplot(
    x='workingday',
    y='count',
    data=day_df,
    ax=axes[0])
axes[0].set_title('Jumlah Pengguna Sepeda berdasarkan Hari Kerja')
axes[0].set_xlabel('Hari Kerja')
axes[0].set_ylabel('Jumlah Pengguna Sepeda')

# based on holidays
sns.barplot(
  x='holiday',
  y='count',
  data=day_df,
  ax=axes[1])
axes[1].set_title('Jumlah Pengguna Sepeda berdasarkan Hari Libur')
axes[1].set_xlabel('Hari Libur')
axes[1].set_ylabel('Jumlah Pengguna Sepeda')

# based on weekdays
sns.barplot(
  x='weekday',
  y='count',
  data=day_df,
  ax=axes[2])
axes[2].set_title('Jumlah Pengguna Sepeda berdasarkan Hari dalam Seminggu')
axes[2].set_xlabel('Hari dalam Seminggu')
axes[2].set_ylabel('Jumlah Pengguna Sepeda')

plt.tight_layout()
plt.show()
```

###### Based on the bar plots above, it can be concluded that the number of bike users is higher on working days compared to holidays. The highest peak of bike users on working days occurs on Thursday, while on holidays it occurs on Saturday.

#### Pertanyaan Bisnis 3: Bagaimana tren penggunaan sepeda sepanjang tahun, dan di antara tahun 2011 dan 2012, mana yang lebih dominan?

```python
day_df['month'] = pd.Categorical(day_df['month'], categories=
    ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'],
    ordered=True)

monthly_counts = day_df.groupby(by=["month", "year"], observed=True).agg({
    "count": "sum"
}).reset_index()

sns.lineplot(
    data=monthly_counts,
    x="month",
    y="count",
    hue="year",
    palette="rocket",
    marker="o")

plt.title("Jumlah total sepeda yang disewakan berdasarkan Bulan dan tahun")
plt.xlabel(None)
plt.ylabel(None)
plt.legend(title="Tahun", loc="upper right")
plt.tight_layout()
plt.show()
```

###### Based on the line plot above, it can be concluded that the trend of bicycle rentals in 2012 is higher than in 2011. The highest peak of bicycle rentals in 2012 occurred in September, while in 2011 it occurred in August.

#### Pertanyaan Bisnis 4: Bagaimana variabel seperti suhu (temp dan atemp) serta kelembaban mempengaruhi total jumlah pengguna sepeda, baik pengguna Casual maupun Registered?

```python
plt.figure(figsize=(14, 6))

# Scatter plot for 'temp' vs 'count'
plt.subplot(1, 3, 1)
sns.scatterplot(
    x='temp',
    y='count',
    data=day_df,
    alpha=0.5
)
plt.title('Temperature vs Count')

# Scatter plot for 'atemp' vs 'count'
plt.subplot(1, 3, 2)
sns.scatterplot(
    x='atemp',
    y='count',
    data=day_df,
    alpha=0.5
)
plt.title('Feels Like Temperature vs Count')

# Scatter plot for 'humidity' vs 'count'
plt.subplot(1, 3, 3)
sns.scatterplot(
    x='humidity',
    y='count',
    data=day_df,
    alpha=0.5
)
plt.title('Humidity vs Count')

plt.tight_layout()
plt.show()
```

###### Based on the scatter plots above, it can be concluded that the temperature (temp and atemp) has a positive correlation with the total number of bike users. Meanwhile, humidity has a negative correlation with the total number of bike users.

#### Pertanyaan Bisnis 5: Apakah musim berpengaruh terhadap jumlah penyewa sepeda di antara pelanggan Casual dan Registered?

```python
# Mengelompokkan data berdasarkan musim dan menghitung jumlah penggunaan terdaftar dan tidak terdaftar
seasonal_usage = day_df.groupby('season', observed=True)[['registered', 'casual']].sum().reset_index()

plt.figure(figsize=(10, 6))

# Membuat
plt.bar(
    seasonal_usage['season'],
    seasonal_usage['registered'],
    label='Registered',
    color='tab:blue'
)

plt.bar(
    seasonal_usage['season'],
    seasonal_usage['casual'],
    label='Casual',
    color='tab:orange'
)

plt.xlabel(None)
plt.ylabel(None)
plt.title('Jumlah penyewaan sepeda berdasarkan musim')
plt.legend()
plt.show()
```

###### Based on the results of the analysis above using a bar plot, it can be concluded that in all seasons the most enjoyed by bicycle users (Casual and Registered) is autumn (Fall), followed by summer (Summer), winter (Winter), and ending with spring (Spring).
